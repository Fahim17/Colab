```
from pdf2image import convert_from_path
pages = convert_from_path('/content/ebook.pdf', dpi=200)

pg = 1

for page in pages:

	# width, height = page.size 
  
	# # Setting the points for cropped image 
	left = 430
	top = 355
	right = 3265
	bottom = 2300
	page = page.crop((left, top, right, bottom)) 
	# page = plOps.invert(plOps.grayscale(page)) 
	page.save('pages/out'+str(pg)+'.jpg')
	pg+=1
	print(type(page))

	if(pg == 10):
		break
```

```
import os
import img2pdf

totalPage = 5
lstofpage = []
for i in range(1,totalPage):
  lstofpage.append(('pages/out'+str(i)+'.jpg'))



with open("output.pdf", "wb") as f:
    # f.write(img2pdf.convert([i for i in os.listdir('/pages') if i.endswith(".jpg")]))
    f.write(img2pdf.convert(lstofpage))
```
